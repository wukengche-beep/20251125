````python
# 這是一個BMI 算法  體重(KG) / 身高 (M) ^2
weight = float(input("請輸入體重"))
height = float(input("請輸入身高"))
bmi = weight / (height / 100) ** 2

print("你輸入的體重是:" + str(weight))
print("你輸入的身高是:" + str(height))
print("計算出的BMI是:" + str(bmi))

# 縮排: tab
# 過濾型的單選題: 嚴格放上面
if bmi >= 25:
    print ("over")
elif bmi >= 18:
    print ("normal")
else:
    print("under")

print("!!!!")
````
