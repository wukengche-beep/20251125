````python
import random
import math

# x^2/4 + y^2/9 = 1
a, b = 2, 3
# 準備一個把它包起來的正方形
w, h = 10, 10

count = 0
times = 1000000

i = 0
while i < times:
    halfw, halfh = w / 2, h / 2
    x = random.uniform(-1*halfw, halfw)
    y = random.uniform(-1*halfh, halfh)
    z = x ** 2 / a ** 2 + y ** 2 / b ** 2
    if z <= 1:
        count = count + 1
    i = i + 1

ratio = count / times
area = ratio * (w * h)
print("猜的是:", area)

real_area = math.pi * a * b
print("我用公式算:", real_area)
````

````python
import urllib.request as req
import json
import pandas as pd

url = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn=36632&geo=TW%2CHK"
f = req.urlopen(url)
content = f.read()
content = json.loads(content)
#型態轉換成 list / dict
danmus = content["data"]["danmu"]
#型態轉換成 pd.DataFrame
df = pd.DataFrame(danmus)
df.to_csv("danmu.csv")
df.to_excel("danmu.xlsx")
````
