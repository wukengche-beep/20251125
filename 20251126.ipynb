````python
import random
import math

# x^2/4 + y^2/9 = 1
a, b = 2, 3
# 準備一個把它包起來的正方形
w, h = 10, 10

count = 0
times = 1000000

i = 0
while i < times:
    halfw, halfh = w / 2, h / 2
    x = random.uniform(-1*halfw, halfw)
    y = random.uniform(-1*halfh, halfh)
    z = x ** 2 / a ** 2 + y ** 2 / b ** 2
    if z <= 1:
        count = count + 1
    i = i + 1

ratio = count / times
area = ratio * (w * h)
print("猜的是:", area)

real_area = math.pi * a * b
print("我用公式算:", real_area)
````
